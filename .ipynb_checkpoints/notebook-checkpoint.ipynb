## Pertanyaan Bisnis
1. **Kapan (jam & hari) permintaan sewa sepeda mencapai puncak, dan bagaimana polanya antar musim?**
2. **Faktor apa yang paling berkorelasi dengan total penyewaan (cnt) pada data harian?**

Notebook ini mencakup: *Data Loading â†’ Data Cleaning â†’ EDA â†’ Visualisasi â†’ Insight & Kesimpulan*.  
Setiap tahap diberi dokumentasi di markdown sesuai pedoman submission.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 100)
%matplotlib inline

day = pd.read_csv("data/day.csv")
hour = pd.read_csv("data/hour.csv")

season_map = {1:"Spring", 2:"Summer", 3:"Fall", 4:"Winter"}
day['season_name'] = day['season'].map(season_map)
hour['season_name'] = hour['season'].map(season_map)

day.head(), hour.head()
```


## Data Cleaning (Quick Checks)
- Mengecek missing values & duplicate
- Tipe data dan rentang nilai


```python
day_info = {
    "rows": len(day),
    "missing": int(day.isna().sum().sum()),
    "duplicates": int(day.duplicated().sum())
}
hour_info = {
    "rows": len(hour),
    "missing": int(hour.isna().sum().sum()),
    "duplicates": int(hour.duplicated().sum())
}
day_info, hour_info
```


## EDA â€” Pertanyaan 1
**Kapan (jam & hari) permintaan sewa sepeda mencapai puncak, dan bagaimana polanya antar musim?**


```python
fig, ax = plt.subplots(figsize=(6,4))
hourly = hour.groupby('hr', as_index=False)['cnt'].mean()
ax.plot(hourly['hr'], hourly['cnt'])
ax.set_title("Rata-rata Penyewaan per Jam")
ax.set_xlabel("Jam (0-23)")
ax.set_ylabel("Rata-rata cnt")
plt.show()

fig, ax = plt.subplots(figsize=(6,4))
byweekday = day.groupby('weekday', as_index=False)['cnt'].mean()
ax.bar(byweekday['weekday'], byweekday['cnt'])
ax.set_title("Rata-rata Penyewaan per Hari (weekday)")
ax.set_xlabel("weekday (0=Sun)")
ax.set_ylabel("Rata-rata cnt")
plt.show()

fig, ax = plt.subplots(figsize=(6,4))
byseason_hour = hour.groupby(['season_name','hr'], as_index=False)['cnt'].mean()
for s, g in byseason_hour.groupby('season_name'):
    ax.plot(g['hr'], g['cnt'], label=s)
ax.set_title("Pola Jam per Musim")
ax.set_xlabel("Jam")
ax.set_ylabel("Rata-rata cnt")
ax.legend()
plt.show()
```


## EDA â€” Pertanyaan 2
**Faktor apa yang paling berkorelasi dengan total penyewaan (cnt) pada data harian?**


```python
corr_cols = ['temp','atemp','hum','windspeed','casual','registered','cnt']
corr = day[corr_cols].corr(numeric_only=True)
corr['cnt'].sort_values(ascending=False)
```


## Insight & Kesimpulan

- **Puncak permintaan** cenderung terjadi pada jam **commute** (pagi & sore). Pola berbeda antar **musim** (Fall/Winter relatif lebih rendah dari Summer).
- Hari kerja vs akhir pekan menunjukkan perbedaan pola (cek grafik weekday).
- Pada data harian, variabel **registered** berkorelasi paling tinggi dengan `cnt`, diikuti oleh **temp/atemp** (suhu lebih hangat â†’ permintaan meningkat).
- Rekomendasi operasional: optimalkan ketersediaan sepeda di jam puncak dan musim dengan permintaan tinggi; perhatikan kebijakan harga/promosi saat cuaca mendukung.
